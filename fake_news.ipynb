```python
# 1. Packages Install Panrom (AI kooda pesa 'requests' thevai)
!pip install fastapi uvicorn nest-asyncio requests
!wget -q https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb
!dpkg -i cloudflared-linux-amd64.deb

from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
import uvicorn
import nest_asyncio
import threading
import time
import requests # API call panrathukku

app = FastAPI()

# 2. CORS Browser Security-ah Allow Panrom
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

class NewsItem(BaseModel):
    text: str

# 🔥 NEENGA FLUTTER-LA VECHIRUNDHA OPENROUTER API KEY IDHU THAAN
OPENROUTER_API_KEY = "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"

# 3. Smart AI Predict Endpoint
@app.post("/predict")
def predict_news(item: NewsItem):
    news_text = item.text

    print("\n" + "="*50)
    print("📥 RECEIVED NEWS:", news_text[:50], "...")
    print("⏳ Asking Smart AI (OpenRouter)...")

    # AI-kitta kekkura kelvi (Prompt)
    headers = {
        "Authorization": f"Bearer {OPENROUTER_API_KEY}",
        "Content-Type": "application/json"
    }

    data = {
        "model": "openai/gpt-3.5-turbo", # Ippo AI news-ah padikka poguthu
        "messages": [
            {
                "role": "system",
                "content": "Analyze this news carefully. If it is true, factual, or verified, start your response with 'REAL'. If it is fake, false, or a rumor, start your response with 'FAKE'. Then provide a 1-sentence reason."
            },
            {"role": "user", "content": f"News: {news_text}"}
        ],
        "temperature": 0.2
    }

    try:
        # AI-kitta irundhu badhil vaangurom
        response = requests.post("https://openrouter.ai/api/v1/chat/completions", headers=headers, json=data)
        result_json = response.json()
        ai_reply = result_json['choices'][0]['message']['content'].strip()

        print("🤖 AI REPLY:", ai_reply)
        print("="*50 + "\n")

        # AI anuppuna text-ah vechu Flutter-ku thevayana madhiri mathurom
        if "FAKE" in ai_reply.upper():
            reason = ai_reply.upper().replace("FAKE", "").strip(" :-.")
            return {"status": "FAKE", "reason": reason}
        elif "REAL" in ai_reply.upper():
            reason = ai_reply.upper().replace("REAL", "").strip(" :-.")
            return {"status": "REAL", "reason": reason}
        else:
            return {"status": "UNKNOWN", "reason": ai_reply}

    except Exception as e:
        print("❌ Error calling AI:", e)
        return {"status": "ERROR", "reason": "Smart AI server-ah connect panna mudiyala."}

nest_asyncio.apply()

# 4. FastAPI-ah background-la run panrom
def run_app():
    uvicorn.run(app, port=8000)

threading.Thread(target=run_app, daemon=True).start()
time.sleep(2)

# 5. Cloudflare Tunnel Start Panrom
print("--------------------------------------------------")
print("🔥 UNGA PUDHU URL KEELA VARUM (Wait for 10 secs)")
print("--------------------------------------------------")
!cloudflared tunnel --url http://127.0.0.1:8000
```
